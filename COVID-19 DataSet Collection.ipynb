```python
import pandas as pd
import shutil
import os
import random
```

```python
File_Path = "covid-chestxray-dataset-master/metadata.csv"
IMAGES_PATH ="covid-chestxray-dataset-master/images"
```

```python
df = pd.read_csv(File_Path)
print(df.shape)
```

```python
df['finding'].unique()
```

```python
find_count = df.groupby('finding')['patientid'].nunique()
find_count
```

```python
sex_count = df.groupby('sex')['patientid'].nunique()
sex_count
```

```python
TARGET_DIR = "Dataset/Covid"

if not os.path.exists(TARGET_DIR):
    os.makedirs(TARGET_DIR)
    print("Created")
```

```python
cnt = 0
row_list = []
for(i,row) in df.iterrows():
    if row["finding"] == "COVID-19" and row["view"] == "PA":
        row_list.append(row)
        filename = row["filename"]
        image_path = os.path.join(IMAGES_PATH,filename)
        image_copy_path = os.path.join(TARGET_DIR,filename)
        shutil.copy2(image_path,image_copy_path)
        cnt += 1
print(cnt)
```

```python
new_df = pd.DataFrame(row_list)  
new_df.head()
 

    
```

```python
covid_df = new_df.loc[:,['url', 'finding','sex','filename']] 
covid_df.reset_index(inplace = True) 
covid_df
```
